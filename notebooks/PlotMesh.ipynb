```python
from superfv import AdvectionSolver, initial_conditions
import matplotlib.pyplot as plt
```

```python
solver = AdvectionSolver(
    ic=initial_conditions.slotted_disk,
    bcx="ic",
    bcy="ic",
    nx=16,
    ny=16,
    nz=1,
    p=3,
    interpolation_scheme="transverse",
    MOOD=True,
)

# make the plot a square
plt.figure(figsize=(6, 6))

plt.plot(solver.X[:, :, 0], solver.Y[:, :, 0], ".", color="k")
plt.plot(
    solver.slab_meshes["x"][0][0][:, :, 0],
    solver.slab_meshes["x"][0][1][:, :, 0],
    ".",
    color="b",
    alpha=0.5,
)
plt.plot(
    solver.slab_meshes["x"][1][0][:, :, 0],
    solver.slab_meshes["x"][1][1][:, :, 0],
    ".",
    color="b",
    alpha=0.5,
)
plt.plot(
    solver.slab_meshes["y"][0][0][:, :, 0],
    solver.slab_meshes["y"][0][1][:, :, 0],
    ".",
    color="r",
    alpha=0.5,
)
plt.plot(
    solver.slab_meshes["y"][1][0][:, :, 0],
    solver.slab_meshes["y"][1][1][:, :, 0],
    ".",
    color="r",
    alpha=0.5,
)

solver.slab_meshes["y"][0][0].min(), solver.slab_meshes["y"][1][0].max()
```
