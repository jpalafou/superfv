```python
from superfv import AdvectionSolver, initial_conditions
import matplotlib.pyplot as plt
```

```python
solver1d = AdvectionSolver(
    ic=initial_conditions.slotted_disk,
    bcx="ic",
    nx=16,
    p=3,
)

plt.figure(figsize=(6, 6))
plt.plot(solver1d.X[:, 0, 0], solver1d.Y[:, 0, 0], ".", color="k")
plt.plot(
    solver1d.slab_meshes["x"][0][0][:, 0, 0],
    solver1d.slab_meshes["x"][0][1][:, 0, 0],
    ".",
    color="b",
    alpha=0.5,
)
plt.plot(
    solver1d.slab_meshes["x"][1][0][:, 0, 0],
    solver1d.slab_meshes["x"][1][1][:, 0, 0],
    ".",
    color="b",
    alpha=0.5,
)
```

```python
solver2d = AdvectionSolver(
    ic=initial_conditions.slotted_disk,
    bcx="ic",
    bcy="ic",
    nx=16,
    ny=16,
    p=3,
)

plt.figure(figsize=(6, 6))
plt.plot(solver2d.X[:, :, 0], solver2d.Y[:, :, 0], ".", color="k")
plt.plot(
    solver2d.slab_meshes["x"][0][0][:, :, 0],
    solver2d.slab_meshes["x"][0][1][:, :, 0],
    ".",
    color="b",
    alpha=0.5,
)
plt.plot(
    solver2d.slab_meshes["x"][1][0][:, :, 0],
    solver2d.slab_meshes["x"][1][1][:, :, 0],
    ".",
    color="b",
    alpha=0.5,
)
plt.plot(
    solver2d.slab_meshes["y"][0][0][:, :, 0],
    solver2d.slab_meshes["y"][0][1][:, :, 0],
    ".",
    color="r",
    alpha=0.5,
)
plt.plot(
    solver2d.slab_meshes["y"][1][0][:, :, 0],
    solver2d.slab_meshes["y"][1][1][:, :, 0],
    ".",
    color="r",
    alpha=0.5,
)
```
