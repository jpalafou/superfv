```python
from superfv import AdvectionSolver, initial_conditions
import matplotlib.pyplot as plt
```

```python
sim1d = AdvectionSolver(
    ic=initial_conditions.slotted_disk,
    bcx="ic",
    nx=16,
    p=3,
)
mesh1d = sim1d.mesh

plt.figure(figsize=(6, 6))
plt.plot(mesh1d.X[:, 0, 0], mesh1d.Y[:, 0, 0], ".", color="k")
plt.plot(
    mesh1d.get_cell_centers("xl")[0][:, 0, 0],
    mesh1d.get_cell_centers("xl")[1][:, 0, 0],
    ".",
    color="b",
    alpha=0.5,
)
plt.plot(
    mesh1d.get_cell_centers("xr")[0][:, 0, 0],
    mesh1d.get_cell_centers("xr")[1][:, 0, 0],
    ".",
    color="b",
    alpha=0.5,
)
```

```python
sim2d = AdvectionSolver(
    ic=initial_conditions.slotted_disk,
    bcx="ic",
    bcy="ic",
    nx=16,
    ny=16,
    p=3,
)
mesh2d = sim2d.mesh

plt.figure(figsize=(6, 6))
plt.plot(mesh2d.X[:, :, 0], mesh2d.Y[:, :, 0], ".", color="k")
plt.plot(
    mesh2d.get_cell_centers("xl")[0][:, :, 0],
    mesh2d.get_cell_centers("xl")[1][:, :, 0],
    ".",
    color="b",
    alpha=0.5,
)
plt.plot(
    mesh2d.get_cell_centers("xr")[0][:, :, 0],
    mesh2d.get_cell_centers("xr")[1][:, :, 0],
    ".",
    color="b",
    alpha=0.5,
)
plt.plot(
    mesh2d.get_cell_centers("yl")[0][:, :, 0],
    mesh2d.get_cell_centers("yl")[1][:, :, 0],
    ".",
    color="r",
    alpha=0.5,
)
plt.plot(
    mesh2d.get_cell_centers("yr")[0][:, :, 0],
    mesh2d.get_cell_centers("yr")[1][:, :, 0],
    ".",
    color="r",
    alpha=0.5,
)
```
