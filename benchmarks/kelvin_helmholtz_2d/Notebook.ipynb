```python
import matplotlib.pyplot as plt

from superfv import OutputLoader, plot_2d_slice

base_path = "/scratch/gpfs/jp7427/out/kelvin-helmholtz/"
```

```python
sim = OutputLoader(f"{base_path}MUSCL-Hancock")
fig, ax = plt.subplots()
plot_2d_slice(sim, ax, "rho")
```

```python
sim1 = OutputLoader(f"{base_path}/MM3/3revs/rtol_1e-2")
sim2 = OutputLoader(f"{base_path}/MM3/3revs/rtol_1e-1")
sim3 = OutputLoader(f"{base_path}/MM7/3revs/rtol_1e-2")
sim4 = OutputLoader(f"{base_path}/MM7/3revs/rtol_1e-1")

rho_min = min(sim.snapshots[-1]["u"][0].min() for sim in [sim1, sim2, sim3, sim4])
rho_max = min(sim.snapshots[-1]["u"][0].max() for sim in [sim1, sim2, sim3, sim4])
```

```python
fig, axs = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(8, 8))

style = dict(variable="rho", vmin=rho_min, vmax=rho_max)
plot_2d_slice(sim1, axs[0, 0], **style)
plot_2d_slice(sim2, axs[1, 0], **style)
plot_2d_slice(sim3, axs[0, 1], **style)
plot_2d_slice(sim4, axs[1, 1], **style)
```
