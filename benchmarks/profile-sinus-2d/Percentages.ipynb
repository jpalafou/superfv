```python
import pandas as pd

from superfv import HydroSolverOutput
```

```python
sim = HydroSolverOutput("/scratch/gpfs/jp7427/superfv/sinus-2d-profiling/cupy/MM3_1rev_rtol_0/N_3072")
```

```python
def build_df(sim):
    def get_cum_time_and_n_calls(cat):
        cum_time = 0.0
        n_calls = 0
        for s in sim.step_history[2:]:
            cum_time += s.timer[cat].cum_time
            n_calls += s.timer[cat].n_calls
        return cum_time, n_calls

    total_cum_time, _ = get_cum_time_and_n_calls("take_step")

    data = []
    for cat in sim.step_history[0].timer.timers.keys():
        cum_time, n_calls = get_cum_time_and_n_calls(cat)
        perc_time = cum_time / total_cum_time * 100
        data.append(dict(cat=cat, n_calls=n_calls, cum_time=cum_time, perc_time=perc_time))

    return pd.DataFrame(data)
        
```

```python
build_df(sim)
```
