```python
def plot_fv(ax, sim):
    ax.set_aspect("equal")

    x_fv, y_fv, _ = sim.mesh.faces
    z_fv = sim.snapshot_history[-1].w[0, :, :, 0].T
    return ax.pcolormesh(x_fv, y_fv, z_fv)


def plot_sd(ax, sim):
    ax.set_aspect("equal")

    x_sd = sim.regular_faces()[0]
    y_sd = sim.regular_faces()[1]
    z_sd = sim.transpose_to_fv(sim.regular_mesh(sim.dm.W_cv))[0]
    return ax.pcolormesh(x_sd, y_sd, z_sd)
```

```python
fig, axs = plt.subplots(5, 5, sharex=True, sharey=True, figsize=(5, 12))
fontsize=8

axs[0,0].set_ylabel("Inviscid")
axs[1,0].set_ylabel(r"$\mathcal{Re}=10000$")
axs[2,0].set_ylabel(r"$\mathcal{Re}=4000$")
axs[3,0].set_ylabel(r"$\mathcal{Re}=2000$")
axs[4,0].set_ylabel(r"$\mathcal{Re}=1000$")

axs[0,0].set_title("MH")
axs[0,1].set_title("FV4")
axs[0,2].set_title("SD4")
axs[0,3].set_title("FV8")
axs[0,4].set_title("SD8")

plot_fv(axs[0,0], sim1)
plot_fv(axs[1,0], sim2)
plot_fv(axs[2,0], sim3)
plot_fv(axs[3,0], sim4)
plot_fv(axs[4,0], sim5)

plot_fv(axs[0,1], sim6)
plot_fv(axs[1,1], sim7)
plot_fv(axs[2,1], sim8)
plot_fv(axs[3,1], sim9)
plot_fv(axs[4,1], sim10)

plot_sd(axs[0,2], sim11)
plot_sd(axs[1,2], sim12)
plot_sd(axs[2,2], sim13)
plot_sd(axs[3,2], sim14)
plot_sd(axs[4,2], sim15)

plot_fv(axs[0,3], sim16)
plot_fv(axs[1,3], sim17)
plot_fv(axs[2,3], sim18)
plot_fv(axs[3,3], sim19)
plot_fv(axs[4,3], sim20)

plot_sd(axs[0,4], sim21)
plot_sd(axs[1,4], sim22)
plot_sd(axs[2,4], sim23)
plot_sd(axs[3,4], sim24)
plot_sd(axs[4,4], sim25)

fig.savefig("viscous_rti.png", dpi=300)
```
